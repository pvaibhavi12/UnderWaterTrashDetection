```python
import os
import glob
from IPython.display import Image, display

def get_latest_predict_dir(base_output_dir='runs/detect'):
    predict_dirs = glob.glob(os.path.join(base_output_dir, 'predict*'))
    if not predict_dirs:
        return None
    # Sort to get the latest (predict, predict-2, predict-3, etc.)
    predict_dirs.sort(key=lambda x: int(x.split('-')[-1]) if '-' in x else 1)
    return predict_dirs[-1]

# Select 5 more unique test images. Assuming test_images variable holds the list of image names.
# We'll pick from the ones that haven't been processed yet based on the previous context.

# Get a fresh list of images in case the previous `test_images` was modified or filtered.
all_test_images = [f for f in os.listdir("dataset/test") if f.endswith(".jpg")]

# Exclude images that were already processed in previous turns (obj0002_frame0000053.jpg and bio0001_frame0000023.jpg)
processed_images = {"obj0002_frame0000053.jpg", "bio0001_frame0000023.jpg"}
remaining_test_images = [img for img in all_test_images if img not in processed_images]

# Pick the next 5 images
images_to_process = remaining_test_images[:5]

for image_file in images_to_process:
    print(f"\nProcessing image: {image_file}")
    source_path = os.path.join("dataset/test", image_file)
    results = model.predict(
        source=source_path,
        save=True,
        conf=0.25
    )

    # Dynamically find the latest predict directory
    latest_predict_dir = get_latest_predict_dir()

    if latest_predict_dir:
        # Construct the possible output filename, considering ultralytics might change the extension
        original_filename_stem = os.path.splitext(os.path.basename(image_file))[0]
        possible_output_filenames = [
            f"{original_filename_stem}.jpg",
            f"{original_filename_stem}.png",
            f"{original_filename_stem}.webp"
        ]

        found_image_path = None
        for possible_name in possible_output_filenames:
            current_path = os.path.join(latest_predict_dir, possible_name)
            if os.path.exists(current_path):
                found_image_path = current_path
                break

        if found_image_path:
            print(f"Displaying: {found_image_path}")
            display(Image(filename=found_image_path))
        else:
            print(f"Could not find predicted image for {image_file} in {latest_predict_dir}")
    else:
        print("No predict directories found.")
```

## Explaining Model Predictions with SHAP and LIME

**SHAP (SHapley Additive exPlanations)** and **LIME (Local Interpretable Model-agnostic Explanations)** are popular techniques used to explain the predictions of complex machine learning models, including deep neural networks.

*   **SHAP** is based on cooperative game theory and aims to explain the output of any machine learning model by computing the contribution of each feature to the prediction. It assigns an importance value (Shapley value) to each feature for a particular prediction.

*   **LIME** explains the predictions of any classifier or regressor by approximating it locally with an interpretable model. For image classification, LIME works by perturbing parts of the input image and observing how the prediction changes, highlighting super-pixels that are most influential to the model's decision.

### Applying SHAP and LIME to Object Detection

Directly applying SHAP or LIME to an end-to-end object detection model like YOLOv5 is not straightforward because these methods are primarily designed for classification or regression tasks where a single output is explained based on input features. For object detection, the output is a set of bounding boxes, class labels, and confidence scores, making the "prediction" more complex.

However, we can adapt these techniques to explain *individual detections* or the model's decision for a *specific class*.

**Proposed approach for LIME (as a starting point):**
1.  **Focus on a single detection:** Pick one bounding box and its associated class prediction from the YOLOv5 output.
2.  **Create a 'pseudo-classifier' function:** This function will take a perturbed image (e.g., an image with parts obscured) and return the confidence score for the *target class* within the region of interest of the original detection. Essentially, we're asking: "If this part of the image changes, how much does the confidence for this specific object class in this location change?"
3.  **Generate explanations:** Use LIME to highlight super-pixels in the image that contribute most positively or negatively to the confidence score of that specific class within that detected bounding box.

Now, let's select an image where the model has made a detection and try to explain that specific detection using LIME. We'll focus on explaining the presence of a 'plastic_bottle' in one of the images.

```python
import cv2
import numpy as np
from lime import lime_image
from skimage.segmentation import mark_boundaries
from PIL import Image as PILImage # Renamed to avoid conflict with IPython.display.Image
import matplotlib.pyplot as plt

# Choose an image for explanation (e.g., one where 'plastic_bottle' was detected)
# We'll use one of the images processed previously, let's pick 'obj1617_frame0000053.jpg'
image_path_for_lime = "dataset/test/obj1617_frame0000053.jpg"

# Load the image using OpenCV
img = cv2.imread(image_path_for_lime)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Perform a prediction on the chosen image to get bounding box and class info
results_for_lime = model.predict(source=image_path_for_lime, save=False, conf=0.25, verbose=False)[0]

# Find a 'plastic_bottle' detection to explain
target_class_name = 'plastic_bottle'
target_class_id = None
# Iterate through model.names to find the class ID for 'plastic_bottle'
for class_id, class_name in model.names.items():
    if class_name == target_class_name:
        target_class_id = class_id
        break

if target_class_id is None:
    print(f"Error: '{target_class_name}' class not found in model names. Current names: {model.names}")
else:
    target_detection = None
    for box_data in results_for_lime.boxes:
        if int(box_data.cls) == target_class_id:
            target_detection = box_data
            break

    if target_detection is None:
        print(f"No '{target_class_name}' detected in {image_path_for_lime}. Please choose another image or adjust confidence.")
    else:
        # Get the bounding box coordinates
        x1, y1, x2, y2 = map(int, target_detection.xyxy[0])
        # LIME expects a classifier function that returns probabilities for all classes
        # We need a function that takes an array of perturbed images and returns a 2D array
        # where each row is the class probabilities for an image.

        # Define a prediction function for LIME
        def predict_fn(images):
            # LIME provides images as numpy arrays, potentially with varying number of channels
            predictions = []
            for i, img_arr in enumerate(images):
                # Ultralytics model.predict expects PIL Image or filepath
                # Ensure image is 3-channel if LIME perturbs it to 1 channel
                if len(img_arr.shape) == 2: # Grayscale
                    img_arr = cv2.cvtColor(img_arr, cv2.COLOR_GRAY2RGB)
                elif img_arr.shape[2] == 4: # RGBA
                    img_arr = cv2.cvtColor(img_arr, cv2.COLOR_RGBA2RGB)

                pil_img = PILImage.fromarray(img_arr.astype(np.uint8))

                # Run prediction on the perturbed image
                yolo_results = model.predict(source=pil_img, save=False, conf=0.25, verbose=False)

                # Create a placeholder array for class probabilities
                class_probs = np.zeros(len(model.names)) # All classes

                # If a detection for our target class exists, use its confidence
                found_target_class_in_perturbed = False
                if yolo_results and yolo_results[0].boxes:
                    for box_data in yolo_results[0].boxes:
                        if int(box_data.cls) == target_class_id:
                            # For LIME, we need confidence per class. We can set it to the confidence of the detected object.
                            # If multiple target class objects are detected, take the highest confidence within the ROI.
                            # For simplicity, we just take the first one found.
                            class_probs[target_class_id] = box_data.conf.item()
                            found_target_class_in_perturbed = True
                            break

                # If the target class is not detected, its probability remains 0 for this perturbed image.
                predictions.append(class_probs)
            return np.array(predictions)

        # Initialize LIME Image Explainer
        explainer = lime_image.LimeImageExplainer()

        # Generate explanation. num_samples controls the number of perturbed samples LIME creates.
        # This can be computationally intensive.
        explanation = explainer.explain_instance(
            img,
            predict_fn,
            top_labels=len(model.names),
            hide_color=0,
            num_samples=1000 # Adjust num_samples for better quality vs speed
        )

        # Get image and mask for the target class
        # top_labels[0] usually refers to the class with the highest probability/confidence in the original image
        # This might not always be our 'target_class_id' if other classes have higher overall confidence
        # It's safer to use the actual target_class_id here for clarity, though LIME might reorder labels internally.
        # Let's ensure we are explaining the specific 'plastic_bottle' class.
        label_to_explain = target_class_id

        temp, mask = explanation.get_image_and_mask(
            label_to_explain,
            positive_only=True,
            num_features=5,    # Number of superpixels to highlight
            hide_rest=True
        )

        # Visualize the explanation
        plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))
        plt.title(f"LIME Explanation for '{model.names[target_class_id]}' in {os.path.basename(image_path_for_lime)}")
        plt.axis('off')
        plt.show()
```

## Understanding Grad-CAM for YOLOv5

**Grad-CAM (Gradient-weighted Class Activation Mapping)** is a technique used to produce visual explanations for predictions made by convolutional neural networks (CNNs). It helps us understand *which parts of an input image* were most important for a model to make a particular prediction. It generates a heatmap that highlights these relevant regions.

### How standard Grad-CAM works (for Image Classification):
1.  **Target Score:** Choose a target class for which you want an explanation (e.g., 'cat' in an image classifier).
2.  **Feature Maps:** Extract the feature maps from a convolutional layer just before the classification head.
3.  **Gradients:** Calculate the gradients of the target class's prediction score with respect to these feature maps.
4.  **Global Average Pooling (GAP) of Gradients:** Average the gradients over the spatial dimensions to get importance weights for each feature map channel.
5.  **Weighted Sum:** Perform a weighted sum of the feature maps, using the gradient weights. This creates a raw heatmap.
6.  **ReLU and Upsampling:** Apply ReLU to the heatmap (to focus on positive contributions) and then upsample it to the original image size. Finally, overlay it on the image to visualize.

### Challenges with Grad-CAM for Object Detection (YOLOv5):

The primary challenge in applying traditional Grad-CAM directly to object detection models like YOLOv5 is that these models don't produce a single, image-level classification score for a given class. Instead, they output a set of bounding boxes, class probabilities, and confidence scores for *multiple potential objects* in an image.

*   **No Single Class Score:** There isn't a direct "plastic_bottle" score for the entire image that we can backpropagate from.
*   **Multiple Detections:** An image can have several detections, each with its own class and confidence.
*   **Complex Output Head:** YOLOv5's detection head is more complex than a simple classification layer, often involving anchors, objectness scores, and coordinate predictions.

### Adapted Approach for Grad-CAM-like Visualizations in YOLOv5:

To apply Grad-CAM concepts to YOLOv5, we need to adapt the target score. The goal becomes: **"What parts of the image are important for *this specific bounding box detecting this specific class*?"**

Here's a conceptual breakdown:

1.  **Select a Target Detection:** First, run YOLOv5 inference and select one specific detected object (a bounding box with a class label and confidence score) that you want to explain. For example, a particular 'plastic_bottle' detection.

2.  **Define a Scalar Target Score:** From this selected detection, we need to extract a single scalar value to backpropagate from. Common choices include:
    *   The **confidence score** of the target class for that specific bounding box.
    *   The **objectness score** for that bounding box if you want to understand *why* the model thinks *something* is there.
    *   (More advanced) A combination of confidence and Intersection Over Union (IOU) with a ground-truth box if available for evaluation.

3.  **Modify Model's Forward Pass (or use Hooks):**
    *   You need to access the intermediate feature maps from a suitable convolutional layer in the YOLOv5 backbone (e.g., a layer from the C3 or SPPF module, or an early P-layer in the FPN neck).
    *   Ensure that the gradients are retained for these feature maps (e.g., using `feature_map.retain_grad()` in PyTorch).

4.  **Calculate Gradients:** Perform a backward pass from the chosen scalar target score (from step 2) with respect to the selected feature maps (from step 3).

5.  **Compute Alpha Weights:** Similar to traditional Grad-CAM, globally average the gradients over the spatial dimensions to get importance weights for each channel of the feature map.

6.  **Generate Heatmap:** Combine the feature maps with these alpha weights, apply ReLU, and then upsample the resulting heatmap to the original image size. This heatmap will show regions important for *that specific detection*.

7.  **Visualize:** Overlay the heatmap onto the original image, possibly along with the bounding box, to visualize the explanation.

### Practical Implementation Notes:

*   **Libraries:** Implementing this from scratch can be complex. Libraries like `pytorch-grad-cam` offer a framework, but adapting them for multi-head object detection models still requires careful coding to correctly hook into the model and define the target score for backpropagation.
*   **YOLOv5 Architecture:** You'll need to understand the YOLOv5 model architecture to pick the most appropriate convolutional layer for extracting feature maps. Layers from the neck (e.g., after the SPPF or C3 modules, or before the detection heads) are often good candidates.
*   **Interpretation:** The generated heatmap will show where the model was 'looking' when it decided that a specific object was present. This can help identify if your model is focusing on relevant features (e.g., the shape of a bottle) or spurious cues (e.g., background textures).

```python
import gradio as gr
from ultralytics import YOLO
from collections import Counter
import numpy as np

# Load trained model
model = YOLO("/content/yolov5/runs/detect/train/weights/best.pt")

def detect(img):

    # Run inference
    results = model(img)

    # Image with bounding boxes
    output_img = results[0].plot()

    # Get detected classes
    boxes = results[0].boxes

    if len(boxes) == 0:
        return output_img, "No trash detected."

    classes = boxes.cls.cpu().numpy()

    # Convert class ids to names
    detected_classes = [
        model.names[int(c)] for c in classes
    ]

    # Count detections
    counts = Counter(detected_classes)

    total_objects = len(detected_classes)

    # Pollution level
    if total_objects <= 2:
        pollution = "Low Pollution"
    elif total_objects <= 5:
        pollution = "Moderate Pollution"
    else:
        pollution = "High Pollution"

    # Build report
    report = "Detected Trash:\n\n"

    for item, count in counts.items():
        report += f"{item}: {count}\n"

    report += f"\nTotal Trash Objects: {total_objects}"
    report += f"\nPollution Level: {pollution}"

    return output_img, report


demo = gr.Interface(
    fn=detect,
    inputs=gr.Image(type="numpy", label="Upload Underwater Image"),
    outputs=[
        gr.Image(label="Detection Result"),
        gr.Textbox(label="Detection Summary")
    ],
    title="🌊 Underwater Trash Detection System",
    description="Upload an underwater image to detect and classify trash."
)

demo.launch(debug=False)
```
